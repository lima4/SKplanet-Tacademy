# 데이터프레임 데이터 추가, 업데이트
- 컬럼추가
```
데이타프레임[새로운컬럼명] = 리스트/넘파이배열/range()
```

- 행추가
```
데이타프레임.loc[새로운행명] = 리스트/넘파이배열/range()
```

````python
data = {"name":["Haidi", "Haidi", "Haidi", "Charles", "Charles"],
        "year":[2014, 2015, 2016, 2015, 2017],
        "points":[1.5, 1.7, 3.6, 2.5, 2.9],
       "penalty" : [1,2,3,4,5]}
df3 = pd.DataFrame(data, columns=["year", "name", "points", 
                                 "penalty"], index=["one", "two", "three", "four", "five"])
df3
````

````python
# 컬럼 추가 'id'라는 컬럼에 동일한 10의 값으로 추가한다.
df3['id'] = 10
df3
````

````python
# 길이 유의 
df3['addr'] = ['LA', 'LA', 'LA', 'Hawaii', 'Hawaii']
df3
````

````python
# 행추가 
df3.loc['six'] = 10
df3
````

````python
# 컬럼에 맞춰서 값 추가하기
df3.loc['six'] = [2017, 'Charles',  3.3, 6, 10,'WC']
df3
````

## 기존값, 수정하는 방법 (업데이트)

데이타프레임.loc[행명, 열명] = 새로운값

데이타프레임.iloc[행인덱스, 열인덱스] = 새로운값


````python
df3.loc['one', 'name'] #haidi를 julia로 변경하기
df3.loc['one', 'name'] = "Julia"
df3
````

````python
df3.iloc[-1, -1] = 'LA'
df3
````

````python
#슬라이싱한 상태에서 값 교체 
df3.loc['one':'three', 'penalty'] = [-500, -1000, -1500]
df3
````

````python
# 특정컬럼의 전체의 값을 변경하고 싶음 
df3.loc[:, 'bonus'] = 300
df3
````

````python
df3['id'] = np.arange(2, df3.shape[0]+2)*10
df3
````

## 행과 열 삭제 

- 열삭제 1
inplace=True 를 지정하면 원본에 바로 적용 
```
데이타프레임명.drop(컬럼명, axis=1, inplace=True)
데이타프레임명.drop([컬럼명1, 컬럼명2 ... ], axis=1, inplace=True)
```
- 열삭제 2
<br> del 데이타프레임명[컬럼명]

- 행삭제 
inplace=True 를 지정하면 원본에 바로 적용 
```
데이타프레임명.drop(행명, axis=0, inplace=True)
데이타프레임명.drop([행명1, 행명2 ... ], axis=0, inplace=True)

```

````python
data = { "name":["Haidi", "Haidi", "Haidi", "Charles", "Charles"],
        "year":[2014, 2015, 2016, 2015, 2017],
        "points":[1.5, 1.7, 3.6, 2.5, 2.9], 
        "penalty":[1, 2, 3, 4, 5] }
df4 = pd.DataFrame( data, 
                    index=["one", "two", "three", "four", "five"])
df4
````

````python
# 간단하게 del로 컬럼 삭제하기
del df4['penalty']
df4
````

````python
#axis=1은  열 axis=1은 행 
#inplace=True 는 원본에서도 삭제 
df4.drop('year', axis=1, inplace = True)
df4
````

````python
# 여러 컬럼 삭제하기
df4.drop(["name", 'points'], axis=1)
````

## 데이터프레임의 복사
- 깊은 복사 VS 얕은 복사
- 얕은 복사란?
: 복사본이나 원본이 같은 메모리번지 사용. <br>
: 서로 데이타가 연결 
- 깊은 복사란?
: 복사본이나 원본이 서로 다른 메모리번지 사용. <br>
: copy() 메서드 이용 

````python
data = { "name":["Haidi", "Haidi", "Haidi", "Charles", "Charles"],
        "year":[2014, 2015, 2016, 2015, 2017],
        "points":[1.5, 1.7, 3.6, 2.5, 2.9], 
        "penalty":[1, 2, 3, 4, 5] }
df5 = pd.DataFrame( data, 
                    index=["one", "two", "three", "four", "five"])
````

````python
df6 = df5
````

````python
# 얕은 복사 방식으로 복제본 생성 
df6 = df5.loc['one':'three', 'name':'year']
df6
````

````python
df6.loc['one', 'name'] = '홍길동'
df6
````

````python
# df6이 변한다고 바뀌지 않는다.
df5
````

````python
#깊은 복사 방식으로 복사본 생성
df7 = df5.copy()
df7
````

````python
df7.iloc[-1, -1] = 500
df7
````

````python
# 원본은 변경 없음
df5
````

# 퀴즈

```
1) df 데이타프레임에서 'Asia' 만 행 추출하여 새로운 데이타프레임으로 저장하여라
2) 1번에서 생성한 데이타프레임의 총 행수는?
3) 1번에서 생성한 데이타프레임에서 year 년도가 (1970~1979)년도의 행만 출력하여라
4) 3번의 결과행을 2번의 데이타프레임에서 제외한후 출력하여라
5) 1번의 데이타프레임에서 gdpPercap 컬럼값이 가장 높은 나라의 정보를 출력하여라 
6) 1번의 데이타프레임에서 lifeExp    컬럼값이 가장 높은 나라는?
7) 1번의 데이타프레임에서 pop 컬럼의 평균값 이상인 나라의 총수는?
```

````python
# 1) df 데이타프레임에서 'Asia' 만 행 추출하여 새로운 데이타프레임으로 저장하여라
df_Asia = df_gap[df_gap['continent']=='Asia']
df_Asia
````

````python
# 2) 1번에서 생성한 데이타프레임의 총 행수는?
df_Asia.shape
````

````python
# 3) 1번에서 생성한 데이타프레임에서 year 년도가 (1970~1979)년도의 행만 출력하여라
df_Asia[(df_Asia['year'] >= 1970) & (df_Asia['year']<= 1979)]
````

````python
# 4) 3번의 결과행을 2번의 데이타프레임에서 제외한후 출력하여라
#quiz[~((quiz < 10) | (quiz >90))]
df_Asia[~((df_Asia['year'] >= 1970) & (df_Asia['year']<= 1979))]
````

````python
# 5) 1번의 데이타프레임에서 gdpPercap 컬럼값이 가장 높은 나라의 정보를 출력하여라 
df_Asia[df_Asia['gdpPercap']== df_Asia['gdpPercap'].max()]
````

````python
# 6) 1번의 데이타프레임에서 lifeExp    컬럼값이 가장 높은 나라는?
df_Asia[df_Asia['lifeExp']==df_Asia['lifeExp'].max()]
````

````python
# 7) 1번의 데이타프레임에서 pop 컬럼의 평균값 이상인 나라의 총수는?


df_Asia[df_Asia['pop']>=df_Asia['pop'].mean()]['country'].unique()
````
