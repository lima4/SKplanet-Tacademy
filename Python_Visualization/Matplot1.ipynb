```python
# 2행, 2열 만들기
# 그래프 변수 설정
fig = plt.figure()

names = ['group_a', 'group_b', 'group_c']
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [80, 40, 100]

# 2행 2열의 n번째 
plt.subplot(221)
plt.plot(names, values1)
plt.subplot(222)
plt.plot(names, values2)
plt.subplot(223)
plt.plot(names, values3)
plt.subplot(224)
plt.plot(names, values4)

plt.show()
```

# Subplot에 데이터 삽입하기 1

- plt.subplot(행열순서)
- plt.plot(xData, yData, label=LabelName)

```python
names = ['group_a', 'group_b', 'group_c']
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]

plt.figure(figsize=(9,3))
plt.subplot(131)
plt.plot(names, values1)

plt.subplot(132)
plt.plot(names, values2)

plt.subplot(133)
plt.plot(names, values3)
# 제목
plt.suptitle('1*3 subplot')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values) # barplot
plt.subplot(132, facecolor='b')
plt.scatter(names, values, c='w') # scatterplot
plt.subplot(133)
plt.plot(names, values) # plot
plt.suptitle('Categorical Plotting') 
plt.show()
```

## Subplot 생성하기2

- 서브플랏 레이아웃 선언 
    - 변수=plt.subplots(행, 열)
- 서브플랏 데이타 그래프는 ? 
    - 변수[i, j].plot(xData, yData)
    
    

```python
# 2행 2열 

# x 데이타 
names = ['group_a', 'group_b', 'group_c']
# y 데이타 
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [80, 40, 100]

# 그래프 전체 변수 설정 
fig = plt.figure()

# 서브플랏리스트 변수 정의 부분 
# 그래프전체변수, 서브플랏리스트 = plt.subplots(row, column)
fig, ax_lst = plt.subplots(2, 2)

ax_lst[0,0].plot(names, values1)
ax_lst[0,1].plot(names, values2)
ax_lst[1,0].plot(names, values3)
ax_lst[1,1].plot(names, values4)

plt.show()
```

```python
# 3행 1열 

# x 데이타 
names = ['group_a', 'group_b', 'group_c']
# y 데이타 
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [80, 40, 100]

# 그래프 전체 변수 설정 
fig = plt.figure()

# 서브플랏리스트 변수 정의 부분 
# 그래프전체변수, 서브플랏리스트 = plt.subplots(row, column)
fig, ax_lst = plt.subplots(3, 1)

# 서브플랏리스트명[행인덱스]
# 서브플랏리스트명[행인덱스, 열인덱스] X
ax_lst[0].plot(names, values1)
ax_lst[1].plot(names, values2)
ax_lst[2].plot(names, values3)
```

```python
# 1행 3열 

# x 데이타 
names = ['group_a', 'group_b', 'group_c']
# y 데이타 
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [80, 40, 100]

# 그래프 전체 변수 설정 
fig = plt.figure()

# 서브플랏리스트 변수 정의 부분 
# 그래프전체변수, 서브플랏리스트 = plt.subplots(row, column)
fig, ax_lst = plt.subplots(1,3)

# 서브플랏리스트명[컬럼인덱스]
# 서브플랏리스트명[행인덱스, 열인덱스] X
ax_lst[0].plot(names, values1)
ax_lst[1].plot(names, values2)
ax_lst[2].plot(names, values3)

plt.show()
```

# Subplot에 데이터 삽입하기 2

- figure 생성
- figureVar, subplotVar=plt.subplots(rowNum , columnNum)
- subPlot[xIndex,yIndex].plot(xData,yData,label=labelName)

```python
names = ['group_a', 'group_b', 'group_c']
values1 = [1, 10, 100]
values2 = [100, 50, 10]
values3 = [100, 40, 100]
values4 = [10, 100, 50]

fig = plt.figure(figsize=(9,3))
fig, ax = plt.subplots(2,2)
ax[0,0].plot(names, values1)
ax[0,1].plot(names, values2)
ax[1,0].plot(names, values3)
ax[1,1].plot(names, values4)
plt.suptitle('Multi Subplots')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

fig11 = plt.figure()
fig11, ax2 = plt.subplots(1,3)
ax2[0].bar(names, values)
ax2[1].plot(names, values)
ax2[2].scatter(names, values)
plt.suptitle('Multi Subplots')
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

fig12 = plt.figure()
fig12, ax3 = plt.subplots(3,1)
ax3[0].bar(names, values)
ax3[1].plot(names, values)
ax3[2].scatter(names, values)
plt.suptitle('Multi Subplots')
plt.show()
```
